```python
iris_data = sns.datasets.load_iris()
iris_df=pd.DataFrame(data=iris_data.data,columns=iris_data.feature_names)
iris_df["species"]=iris_data.target
print(iris_df)

```

```python
sb.pairplot(data=iris_df,hue="species")
plt.show()
```

```python
sns.violinplot(data=iris_df, x='species', y='sepal length (cm)')
plt.show()
```

```python
sns.swarmplot(data=iris_df, x='species', y='sepal length (cm)')
plt.show()
```

```python
import seaborn as sns

sns.kdeplot(data=iris_df['sepal length (cm)'])
```

```python

grid = sns.FacetGrid(data=iris_df, col='species')
grid.map(sns.histplot, 'sepal length (cm)')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.step(iris_df.index, iris_df['sepal length (cm)'])
plt.show()
```

```python
plt.hexbin(iris_df['sepal length (cm)'], iris_df['sepal width (cm)'], gridsize=10, cmap='Blues')
plt.colorbar()
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Hexbin Plot of Sepal Length vs Sepal Width')
plt.show()
```

```python
sns.kdeplot(data=iris_df, x='sepal length (cm)', hue='species', fill=True)
plt.show()
```

```python
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

pd.plotting.scatter_matrix(iris_df.drop('species', axis=1), figsize=(8, 8))
plt.show()
```

```python
sns.regplot(x='sepal length (cm)', y='petal length (cm)', data=iris_df)
plt.show()
```
